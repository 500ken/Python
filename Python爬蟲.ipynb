Python在網頁爬蟲收集資料非常方便，常常看到大家使用urllib、requests、Beautifulsoup進行資料收集，

## urllib

```python
import urllib
```

```python
#先看看urllib中有那些類別
help(urllib)
```

```python
#我們需要用到parse類別中的urlparse函式，用原本import urllib的方式，需要打成urllib.parse.urlparse(連結網址)
#因此改用以下的方式import，直接使用urlparse函式
from urllib.parse import urlparse
#import urllib
```

```python
#這幾天天氣比較冷，就找中央氣象局的天氣預報來試試看，
#以預報台北市為例子：中央氣象局首頁>天氣預報>縣市預報>臺北市     
url="http://www.cwb.gov.tw/V7/forecast/taiwan/Taipei_City.htm"
www=urlparse(url)
print(www)
```

從上面的執行解果可以得知urlparse用來解析網址
通訊協定scheme='http', 
網站名稱netloc='www.cwb.gov.tw', 
路徑path='/V7/forecast/taiwan/Taipei_City.htm', 
url查詢參數字串params='', 
query查詢字串，GET參數query='', 
框架名稱fragment='',
另外有時候會有port通訊埠

```python
import urllib.request
response = urllib.request.urlopen(url)  
html = response.read()  
html
```

urllib在網路上找到很多不同的使用法，需要些時間深入研究，先試試看另一個套件

## requests

```python
import requests
```

```python
#用這個套件來獲取中央氣象局台北市天氣預報的網頁原始碼
url='http://www.cwb.gov.tw/V7/forecast/taiwan/Taipei_City.htm'
html=requests.get(url)
html.encoding='utf-8' #網頁編碼為utf-8
# print(html.text) 結果很長，暫時先不執行
```

```python
htmllist=html.text.splitlines() #將抓取到的原始碼每列做分割
htmllist#類型是list,html.text類型是字串
len(htmllist)
```

```python
#檢查這個頁面是否有臺北市的字串
if "臺北市" in html.text: print("GET!")
#利用if跟for來找找看臺北市出現幾次
count=0
for row in htmllist:
    if "今日白天" in row:
        count+=1
print("GET{}!".format(count))
```
