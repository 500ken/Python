# Python繪圖套件-Matplotlib

Matplotlib可以用來製作2D繪圖，常常與NumPy搭配使用，
而其中最常用到的功能是在matplotlib.pyplot中，
因此會常常看到大家import這個套件，並取名叫plt

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#先試試看隨便用兩個list畫個圖，plt.plot(lista,listb)
lista=[1,3,5,7,9]
listb=[2,4,6,8,10]
plt.plot(lista,listb)  
plt.show()
```

```python
#Matplotlib會自動調整圖片大小還有顏色，我們也可以自己宣告設定，像是color顏色，x(y)label座標標題，lw(linewidth)線寬`，title圖表標題
plt.plot(lista,listb,color="blue",lw=4)
plt.title("training")
plt.xlabel("odd")
plt.ylabel("even")
plt.show()
```

```python
#學長傳授，利用指令畫出y=sin(x)圖形
x=np.linspace(0,2.*np.pi,50)        # 建立線性空間x, 其範圍是0到2pi, 共有50個點
y=np.sin(x)                         # 建立y=sin(x)

plt.plot(x,y,marker='.',ms=5)       # 繪圖，資料點用圓圈標示，圓圈大小5
plt.xlabel('x')                     # 給x軸名稱
plt.ylabel('sin(x)')                # 給y軸名稱
plt.title('my plot')                # 給圖標題

plt.show()                          # 顯示該張圖
```

同一個圖面上，也可以畫兩張圖

```python
x=np.linspace(0,2.*np.pi,50)        # 建立線性空間x, 其範圍是0到2pi, 共有50個點
y1=np.sin(x)                         # 建立y=sin(x)       
y2=np.cos(x)                         # 建立y=sin(x)

plt.plot(x,y1,marker='.',ms=5,label="sin(x)")       # 繪圖，資料點用圓圈標示，圓圈大小5
plt.plot(x,y2,marker='x',ms=5,label="cos(x)")       # 繪圖，資料點用圓圈標示，圓圈大小5
plt.legend()
plt.xlabel('x')                     # 給x軸名稱
plt.ylabel('y')                # 給y軸名稱
plt.title('my plot')                # 給圖標題

plt.show()                          # 顯示該張圖
```

```python
#如果想要分開成兩張，並且放置於左右(上下也可)可以用plt.subplot()
plt.figure(figsize=(10,5))  #用來調整圖面大小
plt.suptitle('my figure')  #設定大標題特別注意這裡是suptitle

(1,2,1)

x=np.linspace(0,2.*np.pi,50)
y=np.sin(x)
plt.plot(x,y,marker='o')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('fig1')

plt.subplot(1,2,2)              #(1,2,2): 畫於1X2圖陣列中的第2個

y=np.cos(x)
plt.plot(x,y,marker='o')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('fig2')

plt.show()
```

```python
#除了線形圖，Matplotlib也可以用來畫出長條圖plt.bar(xlist,ylist)
#將第一個練習的線性函數改為長條圖
lista=[1,3,5,7,9]
listb=[2,4,6,8,10]
plt.bar(lista,listb)  
plt.show()
```

Matplotlib中也有圓餅圖可以繪製，能設定的參數也比線、長條圖多了一些
結構為plt.pie(資料[,選擇性參數]

```python
#以下舉個例子來實際畫畫看
#某小學有一年級三個班，分別是28、24、31人，請以圓餅圖畫出各班佔全體學生數統計
labels=["1st","2nd","3rd"]
people=[28,24,31]
colors=["yellow","pink","blue"]
explode=[0,0,0.1]  #explode表示是否要當作凸出區塊，0為不凸出
#以上為需要用串列表示的選擇性參數
plt.pie(people,explode=explode,labels=labels,colors=colors,#換行
       labeldistance=1.2, #標題離圓心距離為半徑幾倍
       autopct="%2.2f%%", #顯示的百分比格式，小數點前幾位及小數點後幾位
       shadow=True, #圖形陰影
       startangle=90, #圓餅圖起始角度，以逆時針計算
       pctdistance=0.5 #百分比數字離圓心距離為半徑幾倍
       )
plt.legend()
plt.axis("equal")  #此參數用來設定xy軸相等顯示為圓形，否則圖案會變為橢圓
plt.show()
```
